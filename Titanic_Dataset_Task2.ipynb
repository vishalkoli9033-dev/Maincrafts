Start Cleaning of data..
used pandas methods are->
1)df.isnull()-->True if Null else give false..

```python
print(df.isnull())
```

Give number of null values in each column..

```python
print(df.isnull().sum())
```

fill all Null values by mean of complete age column.. 

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

recheck all columns..

```python
print(df.isnull().sum())
```

use maximum replated value in place of null.

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

recheck all columns..

```python
print(df.isnull().sum())
```

Cabin column is useless so we remove it.

```python
df.drop(columns=["Cabin"], inplace=True)
```

recheck all columns..

```python
print(df.isnull().sum())
```

Who survived more: males or females?
-->Females are servived more than males.

```python
df.groupby("Sex")["Survived"].mean()
```

Did passenger class affect survival chances?
-->Yes

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ["Child", "Teen", "Young Adult", "Adult", "Senior"]
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)
df.groupby("AgeGroup", observed=True)["Survived"].mean()
```

From here we start Visualizing of data by using Seaborn / Matplotlib:

Bar chart of survival by gender.

```python
import matplotlib.pyplot as plt

survival_by_gender = df.groupby("Sex")["Survived"].mean()
plt.figure()
plt.bar(survival_by_gender.index, survival_by_gender.values)
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")
plt.show()
```

Bar chart of survival by class.

```python
survival_by_class = df.groupby("Pclass")["Survived"].mean()
plt.figure()
plt.bar(survival_by_class.index.astype(str), survival_by_class.values)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

Histogram of passenger ages.

```python
plt.figure()
plt.hist(df["Age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.title("Distribution of Passenger Ages")
plt.show()
```
