Visualizations..
Visualization is used for good understanding of data by graphs.
--

Here we use subplots to plot histograms on same figure for grade G1,G2,G3.
--

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))
G1=df["G1"]
plt.subplot(1,3,1)
plt.hist(G1,bins=5,edgecolor="black")
plt.title("G1")
G2=df["G2"]
plt.subplot(1,3,2)
plt.hist(G2,bins=5,edgecolor="black")
plt.title("G2")
G3=df["G3"]
plt.subplot(1,3,3)
plt.hist(G3,bins=5,edgecolor="black")
plt.title("G3")
plt.tight_layout()
```

Scatter plot for grade G1,G2,G3.
--

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))
s_time = df["studytime"]
G1 = df["G1"]
G2 = df["G2"]
G3 = df["G3"]
plt.scatter(s_time, G1, label="G1", alpha=0.7)
plt.scatter(s_time, G2, label="G2", alpha=0.7)
plt.scatter(s_time, G3, label="G3", alpha=0.7)
plt.xlabel("Study Time")
plt.ylabel("Grades")
plt.title("Study Time vs Student Performance")
plt.legend()
plt.tight_layout()
plt.show()
```

Average of G1,G2,G3 by comparing Male and Female..
--

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
g1 = df.groupby("sex")["G1"].mean()
plt.bar(g1.index, g1.values, edgecolor="black")
plt.title("Average G1")
plt.xlabel("Sex")
plt.ylabel("Average Score")

plt.subplot(1, 3, 2)
g2 = df.groupby("sex")["G2"].mean()
plt.bar(g2.index, g2.values, edgecolor="black")
plt.title("Average G2")
plt.xlabel("Sex")

plt.subplot(1, 3, 3)
g3 = df.groupby("sex")["G3"].mean()
plt.bar(g3.index, g3.values, edgecolor="black")
plt.title("Average G3")
plt.xlabel("Sex")
plt.tight_layout()
plt.show()
```
